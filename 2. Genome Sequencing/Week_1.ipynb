## Code Challenge: Solve the String Composition Problem.

 Generate the k-mer composition of a string.

Input: An integer k and a string Text.
Output: Compositionk(Text) (the k-mers can be provided in any order).



```python
def composition(text, k, sort=True):
    kmers = list()
    for i in range(len(text)-k+1):
        kmers.append(text[i:i+k])
    if sort:
        return sorted(kmers)
    return kmers
```

```python
k = 5
text = 'CAATCCAAC'

sample_in = (text, k)
sample_out = ['CAATC', 'AATCC', 'ATCCA', 'TCCAA', 'CCAAC']

assert composition(*sample_in, sort=False) == sample_out
```

## String Spelled by a Genome Path Problem. Reconstruct a string from its genome path.

Input: A sequence path of k-mers Pattern1,  ,Patternn such that the last k - 1 symbols of Patterni are equal to the first k-1 symbolsof Patterni+1 for 1  i  n-1.  
Output: A string Text of length k+n-1 such that the i-th k-mer in Text is equal to Patterni (for 1  i  n).



```python
def path_to_genome(path):
    genome = path[0]
    for kmer in path[1:]:
        genome += kmer[-1]
    return genome        
```

```python
path = ['ACCGA', 'CCGAA', 'CGAAG', 'GAAGC', 'AAGCT']
genome = 'ACCGAAGCT'
assert path_to_genome(path) == genome
```

```python
with open('data/dataset_198_3.txt', 'r') as f:
    path = f.read().split()
genome = path_to_genome(path)
print(genome)
```

```python
with open('data/submission_198_3.txt', 'w') as f:
    f.write(genome)
    f.close()
```

## Code Challenge: Solve the Overlap Graph Problem (restated below).

Input: A collection Patterns of k-mers.  
Output: The overlap graph Overlap(Patterns), in the form of an adjacency list. (You may return the nodes and their edges in any order.)

```python
from itertools import combinations

def overlap_graph(patterns):
    adjacencies = list()
    for (i, kmer_i), (j, kmer_j) in combinations(enumerate(patterns), 2):
        if kmer_i[1:] == kmer_j[:-1]:
            adjacencies.append((i, j))
        if kmer_j[1:] == kmer_i[:-1]:
            adjacencies.append((j, i))
    return adjacencies

def list_to_dict(lst):
    """
    Convert a list of tuples to a dictionary. All values are stored in lists,
    even if there is only one value for a particular key.

    Parameters:
        lst (list): A list of tuples.

    Returns:
        dict: A dictionary where the first element of each tuple is the key
              and the second element is the value. All values are stored in
              lists, even if there is only one value for a particular key.
    """
    d = {}
    for k, v in lst:
        if k in d:
            d[k].append(v)
        else:
            d[k] = [v]
    return d


def display_graph(patterns, adjacencies):
    adjacencies = list_to_dict(adjacencies)
    edges = [
        f'{patterns[i]}: {" ".join([patterns[j] for j in adjacencies[i]])}'
        for i in adjacencies.keys()
    ]
    return '\n'.join(edges)
```

```python
patterns = ['ATGCG', 'GCATG', 'CATGC', 'AGGCA', 'GGCAT', 'GGCAC']
adjacencies = overlap_graph(patterns)
print(display_graph(patterns, adjacencies))
```

```python
with open('data/dataset_198_10.txt', 'r') as f:
    patterns = f.read().split()

adjacencies = overlap_graph(patterns)
graph = display_graph(patterns, adjacencies)
# print(graph)
```

## Code Challenge: Solve the De Bruijn Graph from a String Problem.
 
Input: An integer k and a string Text.  
Output: DeBruijnk(Text), in the form of an adjacency list.

```python
from collections import defaultdict

def build_debruijn_graph(text, k):
    """
    Builds a De Bruijn graph from a string of text and a k-mer length.

    Args:
        text: A string of text from which to construct the De Bruijn graph.
        k: An integer representing the k-mer length.

    Returns:
        A dictionary representing the De Bruijn graph, where the keys are strings
        of length k-1 and the values are lists of strings of length k representing
        the outgoing edges.
    """
    # Generate a list of nodes by calling the composition function with k-1.
    nodes = composition(text, k-1, False)
    
    # Create an empty dictionary with default values of empty lists.
    edges = defaultdict(list)
    
    # Iterate over the k-mers in the text and add their edges to the edges dictionary.
    for i in range(len(text)-k+1):
        edges[nodes[i]].append(nodes[i+1])
    
    # Return the edges dictionary as the final De Bruijn graph representation.
    return edges

def sort_dict_keys_and_values(d):
    """
    Sorts the keys and values of a dictionary alphabetically.

    Args:
        d: A dictionary with string keys and lists of strings as values.

    Returns:
        A new dictionary with the same values as d, but with keys and values sorted alphabetically.
    """
    sorted_dict = {}
    for key in sorted(d.keys()):
        sorted_dict[key] = sorted(d[key])
    return sorted_dict

def display_debruijn_graph(graph):
    """
    Creates a string representation of a De Bruijn graph.

    Args:
        graph: A dictionary representing the De Bruijn graph, where the keys are strings
               of length k-1 and the values are lists of strings of length k representing
               the outgoing edges.

    Returns:
        A string representation of the De Bruijn graph, where each line represents a node
        in the graph and its outgoing edges in the format "node: edge1 edge2 ... edgeN".
    """
    # Create a list comprehension that iterates over the items of the graph dictionary.
    # For each item, create a string representation of the node and its outgoing edges
    # in the format "node: edge1 edge2 ... edgeN".
    graph = sort_dict_keys_and_values(graph)
    node_strings = [
        f'{key}: {" ".join(values)}'
        for key, values in graph.items()
    ]
    
    # Join the node strings with newline characters to create the final string representation
    # of the De Bruijn graph.
    return '\n'.join(node_strings)
```

```python
print(display_debruijn_graph(build_debruijn_graph("AAGATTCTCTAAGA", 4)))
```

```python
with open('data/dataset_199_6.txt', 'r') as f:
    data = f.read().splitlines()
    k = int(data[0])
    text = data[1]

graph = build_debruijn_graph(text, k)

with open('data/submission_199_6.txt', 'w') as f:
    f.write(display_debruijn_graph(graph))
```

## DeBruijn Graph from k-mers Problem: Construct the de Bruijn graph from a set of k-mers.

Input: A collection of k-mers Patterns.  
Output: The adjacency list of the de Bruijn graph DeBruijn(Patterns).  
Code Challenge: Solve the de Bruijn Graph from k-mers Problem.

```python
def prefix(kmer):
    """
    Returns the prefix of a k-mer.

    Args:
        kmer: A string representing a k-mer.

    Returns:
        The prefix of the input k-mer, which is the substring of the k-mer
        from the first character up to the second-to-last character.
    """
    return kmer[:-1]


def suffix(kmer):
    """
    Returns the suffix of a k-mer.

    Args:
        kmer: A string representing a k-mer.

    Returns:
        The suffix of the input k-mer, which is the substring of the k-mer
        from the second character up to the last character.
    """
    return kmer[1:]


def debruijn_graph_from_kmers(kmers):
    """
    Constructs a De Bruijn graph from a list of k-mers.

    Args:
        kmers: A list of strings representing k-mers.

    Returns:
        A dictionary representing the De Bruijn graph, where the keys are strings
        representing the nodes of the graph and the values are lists of strings
        representing the outgoing edges from each node.
    """
    # Create an empty dictionary with default values of empty lists.
    edges = defaultdict(list)

    # Iterate over the k-mers in the input list and add their edges to the edges dictionary.
    for kmer in kmers:
        edges[prefix(kmer)].append(suffix(kmer))

    # Return the edges dictionary as the final De Bruijn graph representation.
    return edges
```

```python
kmers = [
    'GAGG',
    'CAGG',
    'GGGG',
    'GGGA',
    'CAGG',
    'AGGG',
    'GGAG',
]

graph = debruijn_graph_from_kmers(kmers)
print(display_debruijn_graph(graph))
```

```python
with open('data/dataset_200_8.txt', 'r') as f:
    data = f.read().split()

graph = debruijn_graph_from_kmers(data)
# print(display_debruijn_graph(graph))
```

```python
with open('data/submission_200_8.txt', 'w') as f:
    f.write(display_debruijn_graph(graph))
```

```python
def is_universal(binary_string, k):
    kmers = composition(binary_string, k)
    unique_kmers = set(kmers)
    if len(unique_kmers) == len(kmers) and len(kmers) == 2 ** k:
        return True
    return False
```
